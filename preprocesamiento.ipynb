# Preprocesamiento

```python
X=df.iloc[:,0:-1]
# columna que expresa si la estadia es alrga o no (0 = no y 1 = si)
y= np.where(df.iloc[:,-1]<=8,0,1)
```

```python
X.drop(['patientid'],axis=1,inplace=True)#''Available Extra Rooms in Hospital','staff_available',gender','Insurance','Ward_Facility_Code''doctor_name','Admission_Deposit'
```

## Tranformo y segun lo pedido.

### Columnas para transformacion

```python
#columnas nominales
ohe_col=['Department', 'Ward_Facility_Code','doctor_name', 'Age', 
     'Type of Admission', 'health_conditions','gender']
#columna ordinal
ord=['Severity of Illness']
#columnas binarias
binar = ['Insurance']
```

### Ordinal

```python
gravedad_enfer = ['Extreme','Moderate','Minor']
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[gravedad_enfer],dtype='uint8')
```

```python
X['Severity of Illness'] = oe.fit_transform(X[['Severity of Illness']])
```

## LaberlBinarizer

```python
X['Insurance'] = X['Insurance'].map({'Yes':1,'No':0})
```

## OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(dtype='uint8')
# ohe.fit_transform(df_copy[['Department']]).toarray()
```

```python
for col in ohe_col:
    df_dep = pd.DataFrame(ohe.fit_transform(X[[col]]).toarray(),columns=[col+'_'+str(ohe.categories_[0][i])for i in range(len(ohe.categories_[0]))])
    X = pd.concat([X,df_dep],axis=1)
    X.drop(col,axis=1,inplace=True)
```
